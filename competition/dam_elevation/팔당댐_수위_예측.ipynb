```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
```

```python
water_list = sorted(glob('/content/water_data/*.csv'))
df_water = pd.DataFrame()
for data in water_list :
  y = pd.read_csv(data)
  df_water = pd.concat([df_water, y],axis = 0, ignore_index=True)

print(len(df_water))
df_water.tail()
```

```python
rf_list = sorted(glob('/content/rf_data/*.csv'))
df_rf = pd.DataFrame()
for i, data in enumerate(rf_list) :
  y = pd.read_csv(data)
  df_rf = pd.concat([df_rf, y],axis = 0, ignore_index=True)

print(len(df_rf))
df_rf.tail()
```

```python
# df_water 와 df_rf 데이터들이 서로 다른 시점이 있는지 확인
n = 0
for i in range(len(df_water)) :
  if df_water['ymdhm'][i] != df_rf['ymdhm'][i]:
    n+=1
print(f'{n}개의 데이터가 df_water 와 df_rf의 시간과 다릅니다.')
```

```python
# 두 데이터 합치기

df = pd.concat([df_water, df_rf], axis = 1)
print(df_water.shape, df.shape)
```

```python
# 데이터 다운캐스팅
def downcast(df, verbos = True) :
  start_mem = df.memory_usage().sum() / 1024**2
  for col in df.columns :
    dtype_name = df[col].dtype.name
    if dtype_name == 'float64' :
      df[col] = pd.to_numeric(df[col], downcast = 'float')
    elif dtype_name == 'int64' :
      df[col] = pd.to_numeric(df[col], downcast = 'integer')
  end_mem = df.memory_usage().sum() / 1024**2
  percentage = (1 - ((start_mem - end_mem) / start_mem)) * 100
  if verbos : 
    print(f'기존 데이터보다 {percentage:.2f}% 만큼 메모리를 적게 사용')
  
  return df
```
